```python
#6Find sum of all left leaves in a given Binary Tree?
# A Binary Tree Node 
class Node: 

    # Constructor to create a new node 
    def __init__(self, data): 
        self.data = data 
        self.left = None
        self.right = None

# Recursive function that traverse the tree and finds the sum of all left leaves 
def sumLeftLeaves(root): 

    # Base Case 
    if (not root): 
        return 0

    # If left child is a leaf node then add it's value 
    if (root.left != None and not root.left.left and not root.left.right): 
        return root.left.data + sumLeftLeaves(root.right) 

    # Else recur for left and right subtrees 
    return sumLeftLeaves(root.left) + sumLeftLeaves(root.right) 

# Driver Code 
if __name__ == '__main__': 

    """ Let us construct below tree 
            20 
            / \ 
            9 49 
            / \ / \ 
        5 12 23 52"""
    
    root = Node(20) 
    root.left = Node(9) 
    root.right = Node(49) 
    root.left.left = Node(5) 
    root.left.right = Node(12) 
    root.right.left = Node(23) 
    root.right.right = Node(52) 

    print("Sum of all left leaves is",sumLeftLeaves(root))
```

```python
# 3) Perform In-order traversal
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def preorder_traversal(root):
    """
    Performs pre-order traversal of a binary tree.
    """
    if root is None:
        return []
    result = []
    result.append(root.val)
    result += preorder_traversal(root.left)
    result += preorder_traversal(root.right)
    return result

def postorder_traversal(root):
    """
    Performs post-order traversal of a binary tree.
    """
    if root is None:
        return []
    result = []
    result += postorder_traversal(root.left)
    result += postorder_traversal(root.right)
    result.append(root.val)
    return result

def inorder_traversal(root):
    """
    Performs in-order traversal of a binary tree.
    """
    if root is None:
        return []
    result = []
    result += inorder_traversal(root.left)
    result.append(root.val)
    result += inorder_traversal(root.right)
    return result

# Example usage
# Create a binary tree with height 2
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Perform traversals
print("Pre-order traversal:", preorder_traversal(root)) # Output: [1, 2, 4, 5, 3]
print("Post-order traversal:", postorder_traversal(root)) # Output: [4, 5, 2, 3, 1]
print("In-order traversal:", inorder_traversal(root)) # Output: [4, 2, 5, 1, 3]
```

```python
#5 Implement BFS (Breath First Search) and DFS (Depth First Search) 
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def bfs_traversal(root):
    """
    Performs BFS traversal of a binary tree.
    """
    if root is None:
        return []
    queue = deque([root])
    result = []
    while queue:
        node = queue.popleft()
        result.append(node.val)
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return result

def dfs_traversal(root):
    """
    Performs DFS traversal of a binary tree.
    """
    if root is None:
        return []
    stack = [root]
    result = []
    while stack:
        node = stack.pop()
        result.append(node.val)
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
    return result

# Example usage
# Create a binary tree with height 2
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Perform BFS and DFS traversals
print("BFS traversal:", bfs_traversal(root)) # Output: [1, 2, 3, 4, 5, 6, 7]
print("DFS traversal:", dfs_traversal(root)) # Output: [1, 2, 4, 5, 3, 6, 7]
```

```python
#2 Find height of a given tree?
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
def tree_height(root):
    if root is None:
        return 0
    left_height = tree_height(root.left)
    right_height = tree_height(root.right)
    return max(left_height, right_height) + 1

# Example usage
# Create a binary tree with height 3
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

print("The height of the binary tree is:", tree_height(root)) # Output: 3
```

```python
#9Find maximum level sum in Binary Tree?
# Python3 program to find maximum level sum in Binary Tree 

# A Binary Tree Node 
class Node: 

    # Constructor to create a new node 
    def __init__(self, data): 
        self.data = data 
        self.left = None
        self.right = None

# Function to find maximum level sum 
def maxLevelSum(root): 

    # Base case 
    if (not root): 
        return 0

    # Initialize result 
    result = root.data 

    # Do Level order traversal keeping track of number of nodes at every level. 
    q = [] 
    q.append(root) 
    while (len(q)): 

        # Get the size of queue when the level order traversal for one level finishes 
        count = len(q) 

        # Iterate for all the nodes in the queue currently 
        sum = 0
        while (count > 0): 
            count -= 1

            # Dequeue an node from queue and add its value to sum 
            temp = q.pop(0) 
            sum += temp.data 

            # Enqueue left and right children of dequeued node 
            if (temp.left): 
                q.append(temp.left) 
            if (temp.right): 
                q.append(temp.right) 

        # Update the maximum node count value 
        result = max(sum, result) 

    return result 

# Driver program to test above function 
root = Node(1) 
root.left = Node(2) 
root.right = Node(3) 
root.left.left = Node(4) 
root.left.right = Node(5) 
root.right.right = Node(8) 
root.right.right.left = Node(6) 
root.right.right.right = Node(7) 

print("Maximum level sum is", maxLevelSum(root))

```
